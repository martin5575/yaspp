```python
df = pd.DataFrame(data=data)
df
```

```python
# Visualisierung der Verteilung der Tordifferenzen (+/-)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="+/-", palette="coolwarm")
plt.title("Häufigkeitsverteilung der Tordifferenzen (+/-)", fontsize=16)
plt.xlabel("Tordifferenz", fontsize=12)
plt.ylabel("Häufigkeit", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Zeitliche Entwicklung der Wahrscheinlichkeiten
time_slots = df["time"].unique()
probabilities = ["-1", "0", "1"]

plt.figure(figsize=(12, 8))
for prob in probabilities:
    plt.plot(
        time_slots,
        df.groupby("time")[prob].mean(),
        label=f"Wahrscheinlichkeit {prob}",
        marker="o",
    )

plt.title("Zeitliche Entwicklung der Wahrscheinlichkeiten", fontsize=16)
plt.xlabel("Zeitintervall", fontsize=12)
plt.ylabel("Wahrscheinlichkeit (%)", fontsize=12)
plt.legend(title="Tordifferenz")
plt.xticks(rotation=45)
plt.grid(axis="both", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# Absolute Häufigkeiten pro Zeitintervall
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="time", y="count", ci=None, palette="viridis")
plt.title("Häufigkeiten pro Zeitintervall", fontsize=16)
plt.xlabel("Zeitintervall", fontsize=12)
plt.ylabel("Häufigkeit", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
