```python
# Visualisierung der Verteilung der Tordifferenzen (+/-)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="+/-", palette="coolwarm", hue="+/-", legend=False)
plt.title("Häufigkeitsverteilung der Tordifferenzen (+/-)", fontsize=16)
plt.xlabel("Tordifferenz", fontsize=12)
plt.ylabel("Häufigkeit", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Zeitliche Entwicklung der Wahrscheinlichkeiten
time_slots = df["time"].unique()
probabilities = ["-1", "0", "1"]

plt.figure(figsize=(12, 8))
for prob in probabilities:
    plt.plot(
        time_slots,
        df.groupby("time")[prob].mean(),
        label=f"Wahrscheinlichkeit {prob}",
        marker="o",
    )

plt.title("Zeitliche Entwicklung der Wahrscheinlichkeiten", fontsize=16)
plt.xlabel("Zeitintervall", fontsize=12)
plt.ylabel("Wahrscheinlichkeit (%)", fontsize=12)
plt.legend(title="Tordifferenz")
plt.xticks(rotation=45)
plt.grid(axis="both", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df2 = df.copy()
df2 = df2[df2["time"] != "00-05"]

# Ensure all time intervals are present
time_intervals = df2["time"].unique()
df2 = df2.set_index(["time", "+/-"]).reindex(pd.MultiIndex.from_product([time_intervals, df["+/-"].unique()], names=["time", "+/-"]), fill_value=0).reset_index()

window_size = 9
df2["-1_MA"] = df2.groupby("+/-")["-1"].transform(lambda x: x.rolling(window=window_size, min_periods=1).mean())
df2["0_MA"] = df2.groupby("+/-")["0"].transform(lambda x: x.rolling(window=window_size, min_periods=1).mean())
df2["1_MA"] = df2.groupby("+/-")["1"].transform(lambda x: x.rolling(window=window_size, min_periods=1).mean())


# Plotting
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plot for -1 probability
for label, grp in df2.groupby("+/-"):
    if label == 0 or label == 1 or label == -1:
        #axes[0].plot(grp["time"], grp["-1"], marker='o', label=f"{label}")
        axes[0].plot(grp["time"], grp["-1_MA"], linestyle='--', label=f"{label} MA")
axes[0].set_title('Probability of -1')
axes[0].set_xlabel('time')
axes[0].set_ylabel('Probability (%)')
axes[0].legend()

# Plot for 0 probability
for label, grp in df2.groupby("+/-"):
    if label == 0 or label == 1 or label == -1:
        #axes[1].plot(grp["time"], grp["0"], marker='o', label=f"{label}")
        axes[1].plot(grp["time"], grp["0_MA"], linestyle='--', label=f"{label} MA")
axes[1].set_title('Probability of 0')
axes[1].set_xlabel('time')
axes[1].set_ylabel('Probability (%)')
axes[1].legend()

# Plot for 1 probability
for label, grp in df2.groupby("+/-"):
    if label == 0 or label == 1 or label == -1:
        #axes[2].plot(grp["time"], grp["1"], marker='o', label=f"{label}")
        axes[2].plot(grp["time"], grp["1_MA"], linestyle='--', label=f"{label} MA")
axes[2].set_title('Probability of 1')
axes[2].set_xlabel('time')
axes[2].set_ylabel('Probability (%)')
axes[2].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
